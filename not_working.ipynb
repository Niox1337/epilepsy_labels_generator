```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Ten_example_Annotation.xlsx")
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)
```

```python
col_counts = {}
new_columns = []
for col in df.columns:
    if col in col_counts:
        col_counts[col] += 1
        new_columns.append(f"{col}_{col_counts[col]}")
    else:
        col_counts[col] = 1
        new_columns.append(col)
df.columns = new_columns
```

```python
print(df.columns.tolist())
```

```python
epilepsy_classification_cols = [
    'epilepsy_class_Unknown',
    'epilepsy_class_Focal',
    'epilepsy_class_Generalised',
    'epilepsy_class_Combined generalised and focal',
    'epilepsy_class_None'
]

comorbidities_cols = [
    'asthma',
    'cerebrovasular accident',
    'chronic obstructive lung disorder',
    'dementia',
    'diabetes',
    'dissociative convulsions',
    'fibromyalgia',
    'hypertension',
    'myocardial infarction'
]

other_medical_conditions_cols = [
    'learning difficulties',
    'pregnant'
]

print("Epilepsy Classification Columns:", epilepsy_classification_cols)
print("Comorbidities Columns:", comorbidities_cols)
print("Other Medical Conditions Columns:", other_medical_conditions_cols)
```

```python
epilepsy_classification_distributions = {}
relevant_epilepsy_rows = df[epilepsy_classification_cols].sum(axis=1) > 0
total_relevant_epilepsy_rows = relevant_epilepsy_rows.sum()

for col in epilepsy_classification_cols:
    epilepsy_classification_distributions[col] = df.loc[relevant_epilepsy_rows, col].sum() / total_relevant_epilepsy_rows

comorbidities_distributions = {}
relevant_comorbidities_rows = df[comorbidities_cols].sum(axis=1) > 0
total_relevant_comorbidities_rows = relevant_comorbidities_rows.sum()


for col in comorbidities_cols:
    comorbidities_distributions[col] = df.loc[relevant_comorbidities_rows, col].sum() / total_relevant_comorbidities_rows

other_medical_conditions_distributions = {}
relevant_other_medical_conditions_rows = df[other_medical_conditions_cols].sum(axis=1) > 0
total_relevant_other_medical_conditions_rows = relevant_other_medical_conditions_rows.sum()

for col in other_medical_conditions_cols:
    other_medical_conditions_distributions[col] = df.loc[relevant_other_medical_conditions_rows, col].sum() / total_relevant_other_medical_conditions_rows
```

```python
excluded_cols = ['document_no', 'hospital_no', 'clinic_date', 'seizure_frequency_number']

valid_cols = [col for col in df.columns if not pd.isna(col)]

feature_cols = epilepsy_classification_cols + comorbidities_cols + other_medical_conditions_cols

target_cols = [col for col in valid_cols if col not in excluded_cols and col not in feature_cols]


print("Feature columns:", feature_cols)
print(len(target_cols), "\nTarget columns:", target_cols)
```

```python
df_features = df[feature_cols].fillna(0)
df_target = df[target_cols].fillna(0)

print("Missing values in df_features after filling:", df_features.isnull().sum().sum())
print("Missing values in df_target after filling:", df_target.isnull().sum().sum())
```

```python
df_target.columns.difference(target_cols)
print(sorted(df_target.columns.to_list()))
print(sorted(target_cols))
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier

base_model = RandomForestClassifier(random_state=42)
multioutput_model = MultiOutputClassifier(base_model, n_jobs=-1)
multioutput_model.fit(df_features, df_target)
```

```python
import numpy as np
import pandas as pd


num_synthetic_samples = 100
synthetic_features = df_features.sample(n=num_synthetic_samples, replace=True, random_state=42)
synthetic_predictions = multioutput_model.predict(synthetic_features)
synthetic_target_df = pd.DataFrame(synthetic_predictions, columns=target_cols)
synthetic_features.reset_index(drop=True, inplace=True)
synthetic_target_df.reset_index(drop=True, inplace=True)
synthetic_df = pd.concat([synthetic_features, synthetic_target_df], axis=1)

display(synthetic_df.head())
```

```python
# Generate a much larger pool of synthetic feature samples by resampling with replacement
num_synthetic_samples_pool = 50000  # Significantly increase the pool size
synthetic_features_pool = df_features.sample(n=num_synthetic_samples_pool, replace=True, random_state=42)

# Use the trained model to predict the target variables for this synthetic feature pool
synthetic_labels_pool_predictions = multioutput_model.predict(synthetic_features_pool)

# Create a DataFrame with the synthetic labels pool
synthetic_labels_pool_df = pd.DataFrame(synthetic_labels_pool_predictions, columns=target_cols)

# Convert DataFrames to a format for easy comparison (e.g., tuples of row values)
original_labels_tuples = [tuple(row) for row in df_target.values]
synthetic_labels_tuples = [tuple(row) for row in synthetic_labels_pool_df.values]

# Filter out synthetic labels that are present in the original labels
distinct_synthetic_labels = [label for label in synthetic_labels_tuples if label not in original_labels_tuples]

# Get unique synthetic labels from the filtered list
unique_distinct_synthetic_labels = pd.DataFrame(list(set(distinct_synthetic_labels)), columns=target_cols)

# Select the desired number of distinct labels (e.g., 1000 or fewer if not enough unique labels)
num_desired_labels = 1000
final_synthetic_labels_df = unique_distinct_synthetic_labels.sample(n=min(num_desired_labels, len(unique_distinct_synthetic_labels)), replace=False, random_state=42)


print(f"Generated {len(final_synthetic_labels_df)} unique and distinct synthetic labels.")
display(final_synthetic_labels_df.head())
```
